```python
import cv2
import numpy as  np
import glob
from google.colab.patches import cv2_imshow
```

```python
image_title='/content/drive/MyDrive/images/L10/letters/c_0.png'
img = cv2.imread(image_title, cv2.IMREAD_GRAYSCALE)
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
magnitude_spectrum = np.asarray(magnitude_spectrum, dtype=np.uint8)
img_and_magnitude = np.concatenate((img, magnitude_spectrum), axis=1)
cv2_imshow(img_and_magnitude)
```

## Iterate over the full folder images in loop


```python
list_images = glob.iglob("/content/drive/MyDrive/images/L10/letters/*")
for image_title in list_images:
    img = cv2.imread(image_title, cv2.IMREAD_GRAYSCALE)
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    magnitude_spectrum = 20*np.log(np.abs(fshift))
    magnitude_spectrum = np.asarray(magnitude_spectrum, dtype=np.uint8)
    img_and_magnitude = np.concatenate((img, magnitude_spectrum), axis=1)

    cv2_imshow(img_and_magnitude)
    

```
